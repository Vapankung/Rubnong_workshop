## State Machine

**Definition of state machine:** A model that explains how a system moves through finite states based on given inputs.

The state machine has the following characteristics:

- Has a fixed set of possible states
- Has a set of inputs that trigger state transitions
- Has a transition function (State changing)
- Has a set of possible outputs

### Example of State Machine

State Machine can be used in various applications such as:

- Light Switch (On–Off) 💡

- Vending machine 🏧

## Example 3: Logic (If-Else)

Logic or `if-else` is one of the most important topics in Python, especially in state machines, since each state needs a **condition** before going to the next state.

**Logic operations:**

- `>` — more than
- `<` — less than
- `==` — equal
- `!=` — not equal
- `and` — both conditions must be true
- `or` — one of the conditions has to be true

### Flowchart



```python
gpa = 2.0

if gpa < 0 or gpa > 4.0:
    print("Invalid Grade")
elif gpa < 1.0:
    print("Status: Retired")
elif gpa < 2.0:
    print("Status: Probation")
else:
    print("Status: Normal")
```

## Example 5: State Machine Using If-Else

A state machine has:

- a **current state**
- an **input / action**
- a **next state**

**Example:** A light switch has two states:
- Light OFF 💡
- Light ON 💡

When the user types `"Flip"`, the switch changes state.

We use `if-else` statements to check the input and change to the next state.  
Moreover, we use `while True` to make the program run continuously until the user types `"quit"`, which breaks the loop and ends the program.

```python
state = "Light OFF"

while True:
    user_input = input("Type 'flip' to switch the light (or 'quit' to stop): ")
    if user_input == "flip":
        if state == "Light OFF":
            state = "Light ON"
        elif state == "Light ON":
            state = "Light OFF"
        print("Current state:", state)
        print("---")
    elif user_input == "quit":
        print("Program ended.")
        break
    else:
        print("Invalid command")
        print("---")
```
