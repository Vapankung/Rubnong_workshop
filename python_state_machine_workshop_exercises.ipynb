# Exercise 3 — If / Elif / Else

**Goal:** Classify student status from GPA.

Rules:
- GPA < 1.00 → Retired
- GPA < 2.00 → Probation
- 2.00 ≤ GPA ≤ 4.00 → Normal
- Otherwise → Invalid Grade


```python
gpa = 2.0

if gpa < 0 or gpa > 4.0:
    print("Invalid Grade")
elif gpa < 1.0:
    print("Status: Retired")
elif gpa < 2.0:
    print("Status: Probation")
else:
    print("Status: Normal")
```

# Example 4 — Loop

**Goal:** Use loop to repeat the actions by iliterating through arrays


```python
# Loop

test_gpas = [0.5, 1.5, 2.0, 4.0, 5.0]

for gpa in test_gpas:
    if gpa < 0 or gpa > 4.0:
        status = "Invalid Grade"
    elif gpa < 1.0:
        status = "Retired"
    elif gpa < 2.0:
        status = "Probation"
    else:
        status = "Normal"

    print(f"GPA = {gpa} -> {status}")
```

# Exercise 5 — State Machine Using If-Else

A **state machine** has:
- current state
- input/action
- next state

Example: A pen has two states:
- `Pen retract`
- `Pen extend`

When the user presses the button, the state changes.


```python
state = "Pen retract"
commands = ["Press button", "Press button", "Press button"]

for user_input in commands:
    print("Input:", user_input)

    if user_input == "Press button":
        if state == "Pen retract":
            state = "Pen extend"
        elif state == "Pen extend":
            state = "Pen retract"

    print("Current state:", state)
    print("---")
```

# Final Challenge — Academic Status State Machine

We will simulate student status across semesters.

States:
- `NORMAL`
- `PROBATION`
- `RETIRED`

Inputs:
- GPA = cumulative average
- GPS = current semester grade

Rules:
- If GPA < 1.00 → `RETIRED`
- If current state is `NORMAL` and GPA < 2.00 → `PROBATION`
- If current state is `PROBATION` and GPS < 2.00 → `RETIRED`
- If current state is `PROBATION` and GPA < 2.00  → `PROBATION`
- Otherwise keep or return to `NORMAL`


```python
NORMAL = "NORMAL"
PRO = "PROBATION"
RETIRED = "RETIRED"

name = input("Enter student name: ")
num_of_sems = float(input("How many semesters do you want to check? "))

current_state = NORMAL

print("\n===== Academic Status Report =====")
print("Student:", name)

for sem in range(1, num_of_sems + 1):
    
    if current_state == RETIRED:
        break
        
    print("\nSemester", sem)
    
    gpa = float(input(f"Enter cumulative GPA semester {sem}: "))
    gps = float(input(f"Enter semester GPS semester {sem}: "))
    
    # Case 1: Student is currently normal
    if current_state == NORMAL:
        
        if gpa < 1.00:
            current_state = RETIRED
            
        elif gpa < 2.00:
            current_state = PRO
            
        else:
            current_state = NORMAL

    # Case 2: Student is currently on probation
    elif current_state == PRO:
        
        if gps < 2.00:
            current_state = RETIRED
                    
        elif gpa < 2.00:
            current_state = PRO
                    
        else:
            current_state = NORMAL

    print(f"GPA (Semester {sem}): {gpa}")
    print(f"GPS (Semester {sem}): {gps}")        
    print(f"Current State (Semester {sem}): {current_state}")


print("\n===== Final Result =====")
print("Student:", name)
print("Final State:", current_state)
```
