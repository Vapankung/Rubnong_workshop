# Exercise 5 — State Machine Using If-Else

A **state machine** has:
- current state
- input/action
- next state

Example: A pen has two states:
- `Pen retract`
- `Pen extend`

When the user presses the button, the state changes.


````python
state = "Pen retract"

while True:
    user_input = input("Type 'Press' to toggle pen state or 'quit' to exit: ")

    if user_input == "Press":
        if state == "Pen retract":
            state = "Pen extend"
        elif state == "Pen extend":
            state = "Pen retract"
    
    elif user_input == "quit":
        print("Program ended.")
        break

    else:
        print("Invalid command")
        print("---")

    print("Current state: ", state)
````

# Final Exercise — Academic Status State Machine

We will simulate student status across semesters.

States:
- `NORMAL`
- `PROBATION`
- `RETIRED`

Inputs:
- GPA = cumulative average
- GPS = current semester grade

Rules:
- If GPA < 1.00 → `RETIRED`
- If current state is `NORMAL` and GPA < 2.00 → `PROBATION`
- If current state is `PROBATION` and GPS < 2.00 → `RETIRED`
- If current state is `PROBATION` and GPA < 2.00  → `PROBATION`
- Otherwise keep or return to `NORMAL`


````python
NORMAL = "NORMAL"
PRO = "PROBATION"
RETIRED = "RETIRED"

name = input("Enter student name: ")
num_of_sems = int(input("How many semesters do you want to check? "))

current_state = NORMAL

print("\n===== Academic Status Report =====")
print("Student:", name)

for sem in range(1, num_of_sems + 1):
    
    if current_state == RETIRED:
        break
        
    print("\nSemester", sem)
    
    gpa = float(input(f"Enter cumulative GPA semester {sem}: "))
    gps = float(input(f"Enter semester GPS semester {sem}: "))
    
    # Case 1: Student is currently normal
    if current_state == NORMAL:
        
        if gpa < 1.00:
            current_state = RETIRED
            
        elif gpa < 2.00:
            current_state = PRO
            
        else:
            current_state = NORMAL

    # Case 2: Student is currently on probation
    elif current_state == PRO:
        
        if gps < 2.00:
            current_state = RETIRED
                    
        elif gpa < 2.00:
            current_state = PRO
                    
        else:
            current_state = NORMAL

    print(f"GPA (Semester {sem}): {gpa}")
    print(f"GPS (Semester {sem}): {gps}")        
    print(f"Current State (Semester {sem}): {current_state}")


print("\n===== Final Result =====")
print("Student:", name)
print("Final State:", current_state)
````

## Challenge 3: Vending Machine State Machine

**Question Statement**

Create a vending machine program using a state machine.

The vending machine has 5 states:


* idle
* vending
* dispensing
* success
* error


The program should start in the `idle` state.

The user can type different commands to move the machine from one state to another.

## Rules

| Input / Action      | Current State | Next State   |
| ------------------- | ------------- | ------------ |
| `insert coin`       | `idle`        | `vending`    |
| `select item`       | `vending`     | `dispensing` |
| `dispense complete` | `dispensing`  | `success`    |
| `fail`              | `vending`     | `error`      |
| `fail`              | `dispensing`  | `error`      |
| `reset`             | `success`     | `idle`       |
| `reset`             | `error`       | `idle`       |

The program should:

1. Print the current state.
2. Ask the user for an action.
3. Change to the next state if the action is valid.
4. Print `"Invalid action"` if the action cannot be used in the current state.
5. Stop the program if the user types `"quit"`.

---

# State Machine Diagram
![Vending_Macjine](pic/State_Machine.png)

````python
state = "idle"

while True:
    print("Current state:", state)

    action = input("Enter action: ").lower()

    if action == "quit":
        print("Program ended.")
        break

    if state == "idle":
        if action == "insert coin":
            state = "vending"
        else:
            print("Invalid action")

    elif state == "vending":
        if action == "select item":
            state = "dispensing"
        elif action == "fail":
            state = "error"
        else:
            print("Invalid action")

    elif state == "dispensing":
        if action == "dispense complete":
            state = "success"
        elif action == "fail":
            state = "error"
        else:
            print("Invalid action")

    elif state == "success":
        if action == "reset":
            state = "idle"
        else:
            print("Invalid action")

    elif state == "error":
        if action == "reset":
            state = "idle"
        else:
            print("Invalid action")

    print("---")
````

## Challenge 4: State machine using dictionary

**Question Statement**

Sometime when we try to use if else, our code might be long and messy. We can use dictionary to store our input and event in the same place to make it easier to call.

**hint**

```python
    state_machine = {
        "current_state_1" : {
            "input_action_1" : "next_state_1",
            "input_action_2" : "next_state_2"
        },

        "current_state_2" : {
            "input_action_1" : "next_state_1",
            "input_action_2" : "next_state_2"
        },
    }
```    

````python
state = "idle"

# State machine table using dictionary
state_machine = {
    "idle": {
        "insert coin": "vending"
    },

    "vending": {
        "select item": "dispensing",
        "fail": "error"
    },

    "dispensing": {
        "dispense complete": "success",
        "fail": "error"
    },

    "success": {
        "reset": "idle"
    },

    "error": {
        "reset": "idle"
    }
}

while True:
    print("Current state:", state)

    action = input("Enter action: ").lower()

    if action == "quit":
        print("Program ended.")
        break

    # Check if action is valid for current state
    if action in state_machine[state]:
        state = state_machine[state][action]
    else:
        print("Invalid action")

    print("---")
````

## Challenge 5: Smart Home Security State Machine (Bell-Gino Challenge)

**Question Statement**

Create a smart home security system program using a state machine.

The security system has 4 states:

* `IDLE`
* `MONITORING`
* `ALERT`
* `ALARM`

The program should start in the `IDLE` state.

The user can type different commands to move the system from one state to another.

---

## Rules

| Input / Action | Current State | Next State   |
| -------------- | ------------- | ------------ |
| `arm`          | `IDLE`        | `MONITORING` |
| `motion`       | `MONITORING`  | `ALERT`      |
| `door`         | `MONITORING`  | `ALERT`      |
| `timeout`      | `ALERT`       | `ALARM`      |
| `disarm`       | `MONITORING`  | `IDLE`       |
| `disarm`       | `ALERT`       | `IDLE`       |
| `disarm`       | `ALARM`       | `IDLE`       |

---

### Additional Requirements

The program should:

1. Print the current state.
2. Ask the user for an action.
3. Change to the next state if the action is valid.
4. Print `"Invalid action"` if the action cannot be used in the current state.
5. Automatically move from `ALERT` to `ALARM` after 5 seconds if the system is not disarmed.
6. Stop the program if the user types `"quit"`.

---

### State Machine Diagram
![Vending_Macjine](pic/Belle-Gino.png)



````python
import time
import threading

state = "IDLE"
alert_start_time = None
event = None
running = True

print("Smart Home Security System Started")
print("Commands: arm, disarm, motion, door, quit")

# -------- INPUT THREAD --------
def get_input():
    global event, running

    while running:
        event = input("Enter event: ").lower()

        if event == "quit":
            running = False
            break


input_thread = threading.Thread(target=get_input, daemon=True)
input_thread.start()

# -------- MAIN LOOP --------
while running:

    # -------- IDLE --------
    if state == "IDLE":
        if event == "arm":
            state = "MONITORING"
            print("System armed. Monitoring started.")
            event = None

    # -------- MONITORING --------
    elif state == "MONITORING":
        if event in ["motion", "door"]:
            state = "ALERT"
            alert_start_time = time.time()
            print("⚠️ ALERT: Suspicious activity detected!")
            event = None

        elif event == "disarm":
            state = "IDLE"
            print("System disarmed.")
            event = None

    # -------- ALERT --------
    elif state == "ALERT":

        if event == "disarm":
            state = "IDLE"
            print("System disarmed before alarm.")
            event = None
            alert_start_time = None

        elif alert_start_time and (time.time() - alert_start_time > 5):
            state = "ALARM"
            print("🚨 ALARM TRIGGERED !!!")
            alert_start_time = None

    # -------- ALARM --------
    elif state == "ALARM":

        if event == "disarm":
            state = "IDLE"
            print("Alarm stopped. System disarmed.")
            event = None
            alert_start_time = None
        else:
            print("🚨🚨 ALARM SOUNDING 🚨🚨")
            time.sleep(1)

    time.sleep(0.5)

print("System stopped.")
````
