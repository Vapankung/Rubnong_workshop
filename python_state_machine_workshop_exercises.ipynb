# Exercise 3 — If / Elif / Else

**Goal:** Classify student status from GPA.

Rules:
- GPA < 1.00 → Retired
- GPA < 2.00 → Probation
- 2.00 ≤ GPA ≤ 4.00 → Normal
- Otherwise → Invalid Grade


````python
gpa = 2.0

if gpa < 0 or gpa > 4.0:
    print("Invalid Grade")
elif gpa < 1.0:
    print("Status: Retired")
elif gpa < 2.0:
    print("Status: Probation")
else:
    print("Status: Normal")
````

````python
# Loop

test_gpas = [0.5, 1.5, 2.0, 4.0, 5.0]

for gpa in test_gpas:
    if gpa < 0 or gpa > 4.0:
        status = "Invalid Grade"
    elif gpa < 1.0:
        status = "Retired"
    elif gpa < 2.0:
        status = "Probation"
    else:
        status = "Normal"

    print(f"GPA = {gpa} -> {status}")
````

# Exercise 4 — Dictionary

**Concept:** A dictionary stores data using `key: value`.

Example:

```python
student = {
    "name": "Pat",
    "year": 3,
    "gpa": 1.5
}
```

This makes student data easier to organize.


````python
student = {
    "name": "Pat",
    "year": 3,
    "gpa": 1.5
}

if student["gpa"] < 1:
    status = "Retired"
elif student["gpa"] < 2:
    status = "Probation"
elif 2 <= student["gpa"] <= 4:
    status = "Normal"
else:
    status = "Invalid Grade"

print(f"Student: {student['name']} | Year: {student['year']} | Status: {status}")
````

# Exercise 5A — State Machine Using If-Else

A **state machine** has:
- current state
- input/action
- next state

Example: A pen has two states:
- `Pen retract`
- `Pen extend`

When the user presses the button, the state changes.


````python
state = "Pen retract"
commands = ["Press button", "Press button", "Press button"]

for user_input in commands:
    print("Input:", user_input)

    if user_input == "Press button":
        if state == "Pen retract":
            state = "Pen extend"
        elif state == "Pen extend":
            state = "Pen retract"

    print("Current state:", state)
    print("---")
````

# Exercise 5B — State Machine Using a Dictionary Table

Instead of many `if-else` statements, we can use a transition table.

The key is:

```python
(input_action, current_state)
```

The value is:

```python
next_state
```


````python
fsm = {
    ("Press button", "Pen retract"): "Pen extend",
    ("Press button", "Pen extend"): "Pen retract",
}

state = "Pen retract"
commands = ["Press button", "Press button", "wrong input", "Press button"]

for user_input in commands:
    print("Input:", user_input)

    if (user_input, state) in fsm:
        state = fsm[(user_input, state)]
    else:
        print("Invalid input")

    print("Current state:", state)
    print("---")
````

# Final Challenge — Academic Status State Machine

We will simulate student status across semesters.

States:
- `NORMAL`
- `PROBATION`
- `RETIRED`

Inputs:
- GPA = cumulative average
- GPS = current semester grade

Suggested rules:
- If GPA < 1.00 → `RETIRED`
- If current state is `NORMAL` and GPA < 2.00 → `PROBATION`
- If current state is `PROBATION` and GPS < 2.00 → `RETIRED`
- If current state is `PROBATION` and GPA ≥ 2.00 and GPS ≥ 2.00 → `NORMAL`
- Otherwise keep or return to `NORMAL`


````python
NORMAL = "NORMAL"
PRO = "PROBATION"
RETIRED = "RETIRED"

name = input("Enter student name: ")
num_of_sems = int(input("How many semesters do you want to check? "))

current_state = NORMAL

print("\n===== Academic Status Report =====")
print("Student:", name)

for sem in range(1, num_of_sems + 1):
    
    if current_state == RETIRED:
        break
        
    print("\nSemester", sem)
    
    gpa = float(input("Enter cumulative GPA: "))
    gps = float(input("Enter semester GPS: "))
    
    # Case 1: Student is currently normal
    if current_state == NORMAL:
        
        if gpa < 1.00:
            current_state = RETIRED
            
        elif gpa < 2.00:
            current_state = PRO
            
        else:
            current_state = NORMAL

    # Case 2: Student is currently on probation
    elif current_state == PRO:
        
        if gps < 2.00:
            current_state = RETIRED
            
        elif gps > 2.00 and gpa < 2.00:
            current_state = PRO
            
        else:
            current_state = NORMAL
            
    print("Current State:", current_state)


print("\n===== Final Result =====")
print("Student:", name)
print("Final State:", current_state)
````
